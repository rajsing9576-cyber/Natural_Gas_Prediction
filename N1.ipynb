```python
import pandas as pd
import numpy as np

# Sample monthly prices
dates = pd.date_range(start='2023-01-31', periods=24, freq='M')

prices = [
    2.8, 2.9, 3.0, 3.1, 3.2, 3.3,
    3.5, 3.7, 3.9, 4.0, 4.2, 4.5,
    4.7, 4.8, 4.9, 5.0, 5.1, 5.2,
    5.0, 4.8, 4.6, 4.5, 4.4, 4.3
]

price_data = pd.DataFrame({
    "Date": dates,
    "Price": prices
})

price_data.set_index("Date", inplace=True)
```

```python
def price_storage_contract(
    injection_dates,
    withdrawal_dates,
    price_data,
    injection_rate,
    withdrawal_rate,
    max_volume,
    storage_cost_per_unit
):
    """
    Returns estimated value of gas storage contract.
    """

    inventory = 0
    total_profit = 0

    for inj_date, wd_date in zip(injection_dates, withdrawal_dates):

        buy_price = price_data.loc[pd.to_datetime(inj_date), "Price"]
        sell_price = price_data.loc[pd.to_datetime(wd_date), "Price"]

        # Volume that can actually be stored
        volume = min(
            injection_rate,
            withdrawal_rate,
            max_volume - inventory
        )

        inventory += volume

        purchase_cost = buy_price * volume

        storage_cost = storage_cost_per_unit * volume

        revenue = sell_price * volume

        cycle_profit = revenue - purchase_cost - storage_cost

        total_profit += cycle_profit

        inventory -= volume

    return total_profit
```

```python
injection_dates = [
    "2023-03-31",
    "2023-06-30"
]

withdrawal_dates = [
    "2023-12-31",
    "2024-03-31"
]

contract_value = price_storage_contract(
    injection_dates=injection_dates,
    withdrawal_dates=withdrawal_dates,
    price_data=price_data,
    injection_rate=1000,
    withdrawal_rate=1000,
    max_volume=5000,
    storage_cost_per_unit=0.10
)

print(f"Contract Value: ${contract_value:,.2f}")
```
